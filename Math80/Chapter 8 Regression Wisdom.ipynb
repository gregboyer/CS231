## 8.1 Examining Residuals

Because residuals are what is "left over" after the model describes the relationaship, they often recceal subtleties that were not clear from the original plot. 

The fundamental assumption in working with a linear model is that the relationship is in fact linear. Sometimes it's much easier to see if that's true from the residuals than from the model itself.
