```python
# 05 Time
#
#
# 05 Time
#
# Demonstrate the use of a generator that indicates
# for each event in access_log the number of seconds
# elapsed since the most recent midnight

# A generator the provides the number of seconds since midnight
# for each event in the log
def get_seconds():
  log = 'access_log'
  for line in open( log ):
    # the timestamp is the string between the square brackets [...]
    timestamp = line.split('[',1)[1].split(']',1)[0]

    # remove the date and timezone from the timestamp
    timestamp = timestamp.split()[0]
    timestamp = timestamp.split(':')

    # convert the hours, minutes, and seconds, in the the seconds since midnight
    seconds = int(timestamp[1])*3600 + int(timestamp[2])*60 + int(timestamp[3])
    yield seconds


# create the generator
elapsed_seconds = get_seconds()

# get the next value from the generator and print it
for seconds in elapsed_seconds:
  print( seconds )
```

```python
'''
CS231, Homework 5, due 09/28/20
Assignment: Demonstrate the use of a generator that indicates for each event in access_log the number
of seconds elapsed since the most recent midnight.
'''

import datetime

# create a sequence of timestamps with only the date and time using a generator expression
# lines are split by spaces and index 3 is selected, then indexes 1 to end of index 3
log_timestamps = ((line.split()[3])[1:] for line in open('access_log'))

''' Note: can't place above generator assignment statement within generator function below b/c will
    cause a printout of only the first event repeated over the printed range of results. '''

# use a generator function to convert the timestamp generator object sequence to date objects
# recent midnights = each date object with time set to zero
def log_date_objects():
    for i in log_timestamps:
        date_obj = datetime.datetime.strptime(i, '%d/%b/%Y:%H:%M:%S') # converts to a date object
        recent_midnight = date_obj.replace(hour=0, minute=0, second=0, microsecond=0)
        seconds_elapsed = int((date_obj - recent_midnight).total_seconds())
        result_str = "has"
        yield i, result_str, seconds_elapsed

# print selection of sequence: iterate through range of date object generator using a list comprehension
for i in [next(log_date_objects()) for _ in range (15)]:
    print ("Event",*i, "seconds since midnight")
```

```python
# Demonstrate the use of a generator that indicates for each event in access_log the number of seconds elapsed since the most recent midnight.

from datetime import datetime
import re

log = open(r'/etc/httpd/logs/access_log','r')

# This function pulls out the timestamp from each line in the log, and converts it into a datetime object.
# It also calculates midnight of the timestamp and subtracts timestamp - midnight
def elapsed():
    time_gen = (re.findall(r'\d\d/\w{3}/\d\d\d\d:\d\d:\d\d:\d\d', line) for line in log)
    for timestamp in time_gen:
        time_obj = datetime.strptime(timestamp[0], '%d/%b/%Y:%H:%M:%S')
        midnight_obj = datetime.strptime(str(timestamp[0])[0:11]+':00:00:00', '%d/%b/%Y:%H:%M:%S')
        yield time_obj - midnight_obj

# Printing the next 20 yields in seconds
for i in range(20):
    print(next(elapsed()).seconds)
```
